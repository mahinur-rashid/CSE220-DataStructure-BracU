```python
def degree_of_matrix(array):
  max=0
  for i in range(len(array[0])):
    sum=0
    for j in range(len(array[0])):
      sum+=array[i][j]
    if sum>max:
      max=sum
      vertex=i
  return vertex,max


v,m=degree_of_matrix(array)
print(f"{l1[v]} has the maximum degree and it is {m}")
```

```python
def degree_of_list(arr):
  count=0
  for i in range(len(arr)):
    temp=arr[i]
    c=0
    while temp!=None:
      c+=1
      temp=temp.next
    if c>count:
      count=c
      index=i

  return count,index

c,i=degree_of_list(l2)
print(f"{l1[i]} vertex has the maximum degree and it is {c}")
```

```python
def transform_dir_mat(arr):
  r,c=arr.shape
  newarr=np.zeros((r,c),dtype=int)
  for i in range(r):
    for j in range(c):
      if arr[i][j]!=0:
        newarr[i][j]=arr[i][j]
        newarr[j][i]=arr[i][j]
  return newarr

na=transform_dir_mat(mat)
print("Directed Matrix")
print(mat)
print("========================")
print("Undirected Matrix")
print(na)
print('========================')
```

```python
def transform_dir_LL(l):
  for j in range(len(l)):
    temp=l[j]

    while temp!=None:
      for i in range(len(l)):
        if temp.destination==l[i].source:
          var=l[i]
          #Checking duplicates in the second while loop
          while var!=None:
              if var.destination==temp.source:
                break
              if var.next==None:
                var.next=Edge(temp.destination,temp.source,temp.weight)
              var=var.next
          break
      temp=temp.next


  return l

list5=transform_dir_LL(ll)

print("Undirected Graph")
for i in range(len(list5)):
  print(f"{destl[i]} available destinations: ")
  temp=list5[i]
  while temp!=None:
    print(f"{temp.destination}({temp.weight})",end="--->")
    temp=temp.next
  print("None")
print("=========================================")
```
